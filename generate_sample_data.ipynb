```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Parameters
days = 365
dates = pd.date_range(start="2024-01-01", periods=days)

# Trend + Seasonality + Noise
np.random.seed(42)
trend = np.linspace(100, 300, days)
seasonality = 20 * np.sin(2 * np.pi * np.arange(days) / 30)
noise = np.random.normal(0, 5, days)

price = trend + seasonality + noise

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Price": price.round(2)
})

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df["Date"], df["Price"])
plt.title("Synthetic Time Series with Trend and Seasonality")
plt.xlabel("Date")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()

# Save file
df.to_csv("forecast_data_365.csv", index=False)

print(df.head())
```
